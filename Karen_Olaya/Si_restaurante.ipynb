```python
from abc import ABC, abstractmethod
from typing import List #Importamos listas para hacer notaciones de objectos
#============================================
#1 Crearemos las clase abstracta persona
#============================================

class Persona(ABC): #Declaramos la super clase persona como abtracta
    def __init__(self, nombre: str) -> None:
        self.nombre = nombre # Corrected indentation

    @property
    def nombre(self) -> str: #GETTER para el nombre de la persona
      return self._nombre #Atributo privado _nombre, devuelve el atributo almacenado en el objeto

    @nombre.setter
    def nombre(self, nombre: str) -> None: #Setter para el nombre de la persona
      #Ahora validamos el nuevo_nombre  ue sea str
      if isinstance(nombre, str) and nombre.strip(): # Corrected: nuevo_nombre.script() -> nombre.strip()
         self._nombre = nombre # Corrected: nuevo_nombre -> nombre
      else:
          #Si no cumple las condiciones del si realizamos esta excepcion para evitar estados invalidos
          raise ValueError("El nombre debe ser cadena de caracteres")

    @abstractmethod # Corrected indentation
    def presentar(self) -> None: #Este metodo debe declararse en todas las clases hijas de persona
        pass # Corrected indentation
#=====================================
#CREAREMOS LAS CLASES HIJAS
#=====================================

class Cliente(Persona):
    def presentar(self) -> None:  #Implementamos el metodo abstracto presentar()
          print(f"El cliente {self.nombre} ha llegado al restaurante")

class Empleado(Persona): #El cliente hereda de persona (ahora es una subclase)
    @abstractmethod
    def trabajar(self) -> None:  #Metodo definido de esta clase,toda subclase de empleado debe implementarlo
      pass

class Mesero(Empleado):
    def presentar(self) -> None:  #Implementamos el metodo abstracto presentar()
           print(f"El Mesero {self.nombre} esta atendiendo a los comensales.")
    def trabajar(self) -> None:
           print(f"El Mesero {self.nombre} esta tomando pedidos.")

class Chef(Persona):
    def presentar(self) -> None:
            print(f"El chef {self.nombre} esta en la cocina.")
    def trabajar(self) -> None:
            print(f"El chef {self.nombre} esta preparando platos deliciosos.")
#========================================================
#CREAREMOS LAS CLASE COCINA(), PERO ESTA NO ES ABC
#========================================================
class Cocina():
  def __init__(self,chefs: List[Chef]) -> None: #Este constructor recibe los objetos Chef y los almacena en una lista
      self.chefs = chefs
  @property
  def chefs(self) -> List[Chef]: #Retornamos las lista privada de _chefs
      return self._chefs
  @chefs.setter
  def chefs(self, lista_chefs: List[Chef]) -> None:
   #Validamos que listas_chefs y que todos sean instancias de class Chef
      if isinstance(lista_chefs, list) and all(isinstance(c, Chef) for c in lista_chefs):
          self._chefs = lista_chefs
      else:
          raise ValueError("Debe proprcionar una lista con elemntos basicos de chef.")
  def operar(self) -> None:
    #Mostramos el trabajo de todos lo chefs
    for chef in self.chefs: #Recorremos la lista chefs
          chef.trabajar() #Llama el metodo trabajar() de cada objeto chef

class Restaurante():
  def __init__(self,nombre: str ) -> None:
    self.nombre = nombre
    self.clientes: List[Cliente] = []
    self.cocina = Cocina([Chef("Manos sucias"), Chef("El chino cochina")])

  @property # Corrected indentation
  def nombre(self) -> str:
    return self._nombre

  @nombre.setter # Corrected indentation
  def nombre(self, nuevo_nombre: str) -> None:
      if isinstance(nuevo_nombre, str) and nuevo_nombre.strip():
           self._nombre = nuevo_nombre
      else:
           raise ValueError("El nombre del restaurante debe se un nuevo nombre valido")

  def agregar_cliente(self,cliente: Cliente) -> None: # Corrected indentation
      if isinstance(cliente, Cliente): #Validamos que cliente sea una instancia de la clase cliente
           self.clientes.append(cliente) #Append agrega el cliente a la lista
      else:
           raise ValueError("El nombre del cliente debe ser un nombre valido")

  def IniciarServicio(self) -> None: # Corrected indentation
       #Simularemos el servio de restaurante
       print(f"\n========Restaurante {self.nombre} iniciando servicio===")
       print(f"\nAtendiendo clientes")
       for cliente in self.clientes: #llamamos al metodo presentar , genera el polimorfismo
           cliente.presentar()
       print("\n COCINA EN OPERACION")
       self.cocina.operar()
#==============
#Main
#=============

def main() -> None:
    #Crearemos el restaurante con el nombre "La Buena Mesa"
    restaurante1 = Restaurante("La Buena Mesa")
    #Agregamos DOS clientes al restaurante
    restaurante1.agregar_cliente(Cliente("JUAN"))
    restaurante1.agregar_cliente(Cliente("MARIA"))
    #SIMULAREMOS INICIO DE OPERACION
    restaurante1.IniciarServicio()


if __name__ == "__main__": # Corrected indentation
    main() # Corrected indentation
```
