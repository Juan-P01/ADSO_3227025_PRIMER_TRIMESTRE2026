```python
class Mueble:
    def __init__(self, nombre, material, color, precio, peso):
        self.nombre = nombre
        self.material = material
        self.color = color
        self.precio = precio
        self.peso = peso

    def mostrar_info(self):
        print("=== Información del Mueble ===")
        print(f"Nombre:   {self.nombre}")
        print(f"Material: {self.material}")
        print(f"Color:    {self.color}")
        print(f"Precio:   ${self.precio}")
        print(f"Peso:     {self.peso} kg")
        print("==============================")


# ===================== OBJETO 1 =====================
mueble1 = Mueble("Silla", "Madera", "Café", 150000, 5)

# 1) Mostrar toda la información
mueble1.mostrar_info()

# 2 y 4) Modificar 3 atributos e indicar cada cambio con un print
print("Modificando atributos del mueble 1...")

mueble1.color = "Negro"
print(f"Color cambiado a: {mueble1.color}")

mueble1.precio = 180000
print(f"Precio cambiado a: ${mueble1.precio}")

mueble1.material = "Metal"
print(f"Material cambiado a: {mueble1.material}")

# 3 y 5) Mostrar los cambios / nuevos valores
print("\n=== Nuevos valores del Mueble 1 ===")
mueble1.mostrar_info()


# ===================== OBJETO 2 =====================
mueble2 = Mueble("Mesa", "Vidrio", "Transparente", 320000, 12)

# 1) Mostrar toda la información
mueble2.mostrar_info()

# 2 y 4) Modificar 3 atributos e indicar cada cambio con un print
print("Modificando atributos del mueble 2...")

mueble2.nombre = "Mesa Grande"
print(f"Nombre cambiado a: {mueble2.nombre}")

mueble2.precio = 400000
print(f"Precio cambiado a: ${mueble2.precio}")

mueble2.peso = 20
print(f"Peso cambiado a: {mueble2.peso} kg")

# 3 y 5) Mostrar los cambios / nuevos valores
print("\n=== Nuevos valores del Mueble 2 ===")
mueble2.mostrar_info()
```
