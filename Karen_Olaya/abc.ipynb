```python
from abc import ABC, abstractmethod
#Crearemos una clase abstracta
class Figura(ABC):
    @abstractmethod
    def calcular_area(self): #Este metodo es obligatorio colocarlo en total las subclases
        pass
#Ahora crearemos una clase hija de figura
class Circulo(Figura):
    def __init__(self, radio): # Added radio as an argument
        self.radio = radio
    def calcular_area(self):
        return (self.radio**2 * 3.1416) # Corrected area calculation for pi
class Cuadrado(Figura):
    def __init__(self, lado):
        self.lado = lado
    def calcular_area(self):
        return({self.lado **2})

class Piramide(Figura):
   def __init__(self, lado):
        self.lado = lado
   def calcular_area(self):
        return({self.lado * self.lado})

class Rectangulo(Figura):
    def __init__(self, lado1 , lado2):
        self.lado1 = lado1
        self.lado2 =  lado2
    def calcular_area(self):
        return({self.lado * self.lado})

class Triangulo(Figura):
    def __init__(self, lado):
        self.lado1 = lado1
        self.lado2= lado2
    def calcular_area(self):
        return({self.lado * self.lado})

# Vamos a ejecutar
def mostrar_area(figura: Figura):
    print(f"Area: {figura.calcular_area():,.2f}")


def main(): # Removed incorrect indentation for main()
   radio1 = float(input("Ingrese el radio del circulo")) # Corrected input function call
   circulo1 = Circulo(radio1)
   mostrar_area(circulo1)


if __name__ == "__main__":
    main()
```
